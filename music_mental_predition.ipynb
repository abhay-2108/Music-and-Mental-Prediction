```python
#Model Building
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X = df.drop(columns=['Anxiety', 'Depression', 'Insomnia', 'OCD'])
y = df[['Anxiety', 'Depression', 'Insomnia', 'OCD']]
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import Binarizer

models = {}
predictions = {}
reports = {}

# Binarize the target variables
threshold_value = y.median().median()  # Use the median of the medians as the threshold
binarizer = Binarizer(threshold=threshold_value)

y_train_binarized = binarizer.fit_transform(y_train)
y_test_binarized = binarizer.transform(y_test)

for i, target in enumerate(y.columns):
    print(f"Training model for {target}...")
    model = RandomForestClassifier(random_state=42, n_estimators=100)
    model.fit(X_train, y_train_binarized[:, i])
    models[target] = model

    # Make Predictions
    y_pred = model.predict(X_test)
    predictions[target] = y_pred

    # Evaluate the Model
    acc = accuracy_score(y_test_binarized[:, i], y_pred)
    print(f"Accuracy for {target}: {acc}")
    reports[target] = classification_report(y_test_binarized[:, i], y_pred)
```

```python
for target, report in reports.items():
    print(f"\nClassification Report for {target}:")
    print(report)
```

```python
# Feature Importance for Each Model
for target, model in models.items():
    feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': model.feature_importances_})
    feature_importances.sort_values(by='Importance', ascending=False, inplace=True)
    print(f"\nTop 5 Features Influencing '{target}':")
    print(feature_importances.head(5))
```
